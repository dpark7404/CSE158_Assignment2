```python
# Function to standardize specific columns in an array
def standardize_columns(array, columns):
    # Extract the columns to be standardized
    data_to_standardize = array[:, columns]
    
    # Initialize the scaler
    scaler = StandardScaler()
    
    # Standardize the selected columns
    standardized_data = scaler.fit_transform(data_to_standardize)
    
    # Replace the original columns with the standardized ones
    array[:, columns] = standardized_data
    
    return array
```

```python
X = []
Y = []

for rating in allRatings:
    missing = False
    # Initialize X_elem as a list
    # X_elem = [1, rating['item_id']]
    X_elem = [1]

    # Handle 'height'
    if 'height' in rating:
        inch = int(rating['height'][-2])  # Convert last two characters to an integer
        feet = int(rating['height'][0]) * 12  # Convert first character to feet
        height_total = feet + inch
        X_elem.append(height_total)
    else:
        X_elem.append(None)
        missing = True

    # Handle 'weight'
    if 'weight' in rating:
        X_elem.append(int(rating['weight'][:-3]))  # Remove "lbs" and convert to int
    else:
        X_elem.append(None)
        missing = True

    if 'size' in rating:
        X_elem.append(rating['size'])
    else:
        X_elem.append(None)
        missing = True

    # Handle 'bust size'
    if 'bust size' in rating:
        # Extract band size (first two characters)
        band_size = int(rating['bust size'][:2])
        X_elem.append(band_size)

        # Map cup size to an ordinal value
        cup_size = rating['bust size'][2:].lower()  # Lowercase for consistent comparison
        cup_mapping = {
            "aa":1,
            "a": 2,
            "b": 3,
            "c": 4,
            "d": 5,
            "dd": 6,
            "d+": 6,
            "ddd/e": 7,
            "f": 8,
            "g":9,
            "h":10,
            "i":11,
            "j":12
        }
        X_elem.append(cup_mapping.get(cup_size, None))  # Use None for unmapped cup sizes
    else:
        X_elem.extend([None, None])  # Add placeholders for band size and cup size
        missing = True


    if not missing:
        # Append the processed X_elem to X
        X.append(X_elem)

        # Append the fit to Y
        if rating['fit'] == 'fit':
            Y.append(1)
        else:
            Y.append(0)

# Convert X and Y to NumPy arrays
X = np.array(X, dtype=object)  # Use dtype=object to handle None values
Y = np.array(Y, dtype=int)
```

```python
def accuracy(predictions, y):
    correct_prediction = 0
    for i in range(len(predictions)):
        if predictions[i] == y[i]:
            correct_prediction += 1
    return correct_prediction / len(predictions)
```

```python
def BER(predictions, y):
    true_positive = 0
    label_positive = 0

    true_neagative = 0
    label_negative = 0

    for i in range(len(predictions)):
        if predictions[i] == y[i]:
            if predictions[i] == 1:
                true_positive += 1
            else:
                true_neagative += 1
        
        if y[i] == 1:
            label_positive += 1
        else:
            label_negative += 1
    TPR = true_positive / label_positive
    TNR = true_neagative / label_negative
    BER = 1 - (0.5 * (TPR + TNR))
    return BER
```

```python
acc1 = accuracy(pred, Y)
ber1 = BER(pred, Y)
```

## Hyperparameter Tuning

```python
Xtrain, Xvalid, Xtest = X[:len(X)//2], X[len(X)//2:(3*len(X))//4], X[(3*len(X))//4:]
ytrain, yvalid, ytest = Y[:len(X)//2], Y[len(X)//2:(3*len(X))//4], Y[(3*len(X))//4:]
```

```python
mod = linear_model.LogisticRegression(C=1, class_weight='balanced')
mod.fit(Xtrain,ytrain)
```

```python
# Define hyperparameter grid
param_grid = {
    'solver': ['lbfgs', 'liblinear', 'saga'],  # Solvers to try
    'C': [0.01, 0.1, 1, 10, 100],             # Regularization strengths
    'max_iter': [100, 500, 1000]              # Iteration limits
}

# Initialize the Logistic Regression model
model = LogisticRegression(class_weight='balanced')

# Set up GridSearchCV
grid_search = GridSearchCV(
    estimator=model,
    param_grid=param_grid,
    cv=5,  # 5-fold cross-validation
    scoring='accuracy',  # Metric to optimize (can be 'f1', 'roc_auc', etc.)
    verbose=1,  # To display progress
    n_jobs=-1   # Use all available cores
)

# Perform the grid search on the training data
grid_search.fit(Xtrain, ytrain)
```

```python
# Best hyperparameters
print("Best hyperparameters:", grid_search.best_params_)

# Use the best model on the validation set
best_model = grid_search.best_estimator_
yvalid_pred = best_model.predict(Xvalid)
valid_accuracy = accuracy_score(yvalid, yvalid_pred)
print("Validation Accuracy:", valid_accuracy)

# Test the final model on the test set
ytest_pred = best_model.predict(Xtest)
test_accuracy = accuracy_score(ytest, ytest_pred)
print("Test Accuracy:", test_accuracy)
```
